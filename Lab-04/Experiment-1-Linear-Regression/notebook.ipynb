# Machine Learning Lab-04: Experiment 1 - Linear Regression

## Objective
Perform Linear Regression using the 'Sales vs TV Marketing' dataset to predict sales based on TV marketing budget.

## Theory
Linear Regression is a statistical method for modeling the relationship between a dependent variable (Y) and one or more independent variables (X). In Simple Linear Regression, we model the relationship between two continuous variables. The goal is to find the best-fitting line that minimizes the sum of squared differences between the observed and predicted values.

Equation: $Y = mx + c$
Where:
- $Y$ is the dependent variable (Sales)
- $x$ is the independent variable (TV Budget)
- $m$ is the slope (coefficient)
- $c$ is the intercept

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, r2_score

# Set style for clean plots
sns.set(style='whitegrid')
```

### 1. Load and Inspect Dataset

```python
# Load the dataset
df = pd.read_csv('sales_tv_marketing.csv')

# Display first 5 rows
print(df.head())

# Check dataset info
print(df.info())
```

### 2. Data Visualization

```python
plt.figure(figsize=(8, 6))
sns.scatterplot(x='TV', y='Sales', data=df)
plt.title('TV Marketing vs Sales')
plt.xlabel('TV Budget')
plt.ylabel('Sales')
plt.show()
```

### 3. Data Preprocessing

```python
# Define features (X) and target (y)
X = df[['TV']]
y = df['Sales']

# Split into training and testing sets (70% train, 30% test)
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

print(f'Training set size: {X_train.shape[0]}')
print(f'Testing set size: {X_test.shape[0]}')
```

### 4. Model Training

```python
# Initialize and train the model
lr_model = LinearRegression()
lr_model.fit(X_train, y_train)

# Creating model parameters
print(f'Intercept: {lr_model.intercept_}')
print(f'Coefficient: {lr_model.coef_[0]}')
```

### 5. Prediction and Visualization

```python
# Make predictions on the test set
y_pred = lr_model.predict(X_test)

# Visualize the regression line
plt.figure(figsize=(8, 6))
sns.scatterplot(x='TV', y='Sales', data=df, label='Data Points')
plt.plot(X_test, y_pred, color='red', linewidth=2, label='Regression Line')
plt.title('Linear Regression Fit')
plt.xlabel('TV Budget')
plt.ylabel('Sales')
plt.legend()
plt.show()
```

### 6. Model Evaluation

```python
# Calculate metrics
mse = mean_squared_error(y_test, y_pred)
r2 = r2_score(y_test, y_pred)

print(f'Mean Squared Error (MSE): {mse:.2f}')
print(f'R-squared Score: {r2:.2f}')
```

## Conclusion
The Linear Regression model was successfully trained on the TV vs Sales dataset. The visualization shows a clear positive correlation between TV marketing budget and sales, which the model captured effectively. The evaluation metrics (MSE and R2 score) confirm the model's performance.
